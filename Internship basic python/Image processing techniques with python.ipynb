```python
# to find the specific colors from image
# Red => red_plane - gray
# Green=>green_plane - gray
# Blue=> blue_plane - gray
# # Cyan=> gray-red_plane
# # Magenta=> gray-green_plane
# Yellow => gray-blue_plane
img5 = plt.imread(r"C:\Users\mk744\OneDrive - Poornima University\Pictures\imgrgb.jpg")
# plt.imshow(img5)
# plt.axis('off')
# plt.show()
```

```python
# Find it's shape
img5.shape
```

```python
gray = cv2.cvtColor(img5, cv2.COLOR_BGR2GRAY)
red_plane = img5[:,:,-1].copy()
green_plane = img5[:,:,-1].copy()
blue_plane = img5[:,:,-1].copy()
yellow_plane = img5[:,:,-1].copy()
cyan = img5[:,:,-1].copy()
magenta = img5[:,:,-1].copy()
```

```python
red_color = cv2.subtract(red_plane, gray)
green_color = cv2.subtract(green_plane, gray)
blue_color = cv2.subtract(blue_plane, gray)
yellow_color = cv2.subtract(gray, blue_plane)
cyan = cv2.subtract(gray, red_plane)
magenta = cv2.subtract(gray, green_plane)
```

```python
def imshow(img5, title, type = '',):
     if type:
          plt.imshow(img5, cmap = type)
     else:
          plt.imshow(img5)
     plt.title(title)
     plt.axis('off')
```

```python
plt.figure(figsize = (18,5))
plt.subplot(3,4,1); imshow(img5, 'Original')
plt.subplot(3,4,2); imshow(gray, 'Gray',  'gray')
plt.subplot(3,4,3); imshow(red_plane,'RP' , 'gray')
plt.subplot(3,4,4); imshow(green_plane,'GP',  'gray')
plt.subplot(3,4,5); imshow(blue_plane,'BP',  'gray')
plt.subplot(3,4,6); imshow(yellow_plane,'YP',  'gray')
plt.subplot(3,4,7); imshow(cyan,'C',  'gray')
plt.subplot(3,4,8); imshow(magenta,'M',  'gray')
plt.subplot(3,4,6); imshow(red_color,'Red',  'gray')
plt.subplot(3,4,7); imshow(green_color,'Green',  'gray')
plt.subplot(3,4,8); imshow(blue_color,'Blue',  'gray')
plt.subplot(3,4,9); imshow(yellow_color,'Yellow',  'gray')
plt.subplot(3,4,10); imshow(magenta,'Magenta',  'gray')
plt.subplot(3,4,11); imshow(cyan,'Cyan',  'gray')
plt.show()
```
