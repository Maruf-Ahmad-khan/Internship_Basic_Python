```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# Reading the files
df_used_cars = pd.read_csv(r"C:\Users\mk744\Downloads\365 data science data sets\Brand,Cars Listings.txt")
```

```python
df_used_cars
```

```python
# Visualisation on Bar Chart
plt.figure(figsize = (9,6))
plt.bar(x = df_used_cars["Brand"],
        height= df_used_cars["Cars Listings"],
        color = "midnightblue")
plt.xticks(rotation = 45, fontsize = 13)
plt.yticks(fontsize = 13)
plt.title("Car Listing by Brand", fontsize = 16, fontweight = "bold")
plt.ylabel("Number of Listings", fontsize = 13)
plt.show()
plt.savefig("Used Cars Bar.png")
```

```python
sns.set_palette('colorblind')
```

```python
# Pie Char Visualisation
plt.figure(figsize = (10,8))
plt.pie(df_used_cars['Cars Listings'],
        labels= df_used_cars['Brand'].values,
        autopct= '%.2f%%',
        textprops= {'size' : 'x-large',
                    'fontweight' : 'bold',
                    'rotation' : '30',
                    'color' : 'w'})
plt.legend(loc="center")
plt.title('Car Listing by Brand', fontsize = 18, fontweight = 'bold', color = 'pink')
plt.show()
plt.savefig("Used Cars Bar.png")
```

```python
# Stack plot Visualisation
colors = ['#011638', "#7e2987"]
labels = ['Brand', 'Car Listings']
sns.set_style("white")
plt.figure(figsize = (9,6))
plt.stackplot(df_used_cars['Brand'],
              df_used_cars['Cars Listings'],
              colors = colors,
              edgecolor = 'none')
plt.xticks(rotation = 45)
plt.legend(labels = labels, loc = 'upper left')
plt.ylabel('Cars Listings', fontsize = 13)
plt.title("Car listings by Brand", fontsize = 14, weight = 'bold')
sns.despine()
plt.show()
```

```python
# Histogram Visualisation
plt.figure(figsize=(9,6))
plt.hist(df_used_cars['Cars Listings'],
         bins = 8,
         color= "#108A99")
plt.title("Cars Listings", fontsize = 14, weight = "bold")
sns.despine()
plt.show()
```

```python
# Scatter plot visualisation
plt.figure(figsize=(9,6))
plt.scatter(df_used_cars['Brand'],
            df_used_cars['Cars Listings'],
            alpha= 0.6,
            c = df_used_cars['Cars Listings'],
            cmap='viridis')
plt.show()
```

```python
# Reading the data in csv
data = pd.read_csv(r"C:\Users\mk744\Downloads\365 data science data sets\Time series data.txt")
```

```python
# to correct the date format
data["new_date"] = pd.to_datetime(data["Date"])
```

```python
data["new_date"]
```

```python
# Line Chart Visualisation
labels = ["S&P 500", "FTSE 100"]
plt.figure(figsize= (20,8))
plt.plot(data["new_date"], data["GSPC500"])
plt.plot(data["new_date"], data["FTSE100"])
plt.title("S&P vs FTSE Returns (2000 - 2010)", fontsize = 14, fontweight = "bold")
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels = labels, fontsize = "large")
plt.show()
```

```python
dataset = data[(data.new_date >= '2008-07-01') &
               (data.new_date <= '2008-12-31')]
```

```python
dataset
```

```python
# labels = ["S&P 500", "FTSE 100"]
plt.figure(figsize= (20,8))
plt.plot(dataset["new_date"], dataset["GSPC500"], color = "midnightblue")
plt.plot(dataset["new_date"], dataset["FTSE100"], color = "crimson")
plt.title("S&P vs FTSE Returns (H2 2008)", fontsize = 14, fontweight = "bold")
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels = labels, fontsize = "large")
plt.show()
```

```python
# scatter plot Visualisation
df_read_estate = pd.read_csv(r"C:\Users\mk744\Downloads\365 data science data sets\Scatter plot.txt")
```

```python
df_read_estate
```

```python
plt.figure(figsize = (12,8))
scatter = plt.scatter(df_read_estate['Area (ft.)'],
            df_read_estate['Price'],
            alpha=0.6,
            c = df_read_estate['Building Type'],
            cmap = 'viridis')
plt.legend(*scatter.legend_elements(),
           loc = "upper left",
           title = "Building Type")
plt.title("Relationship between Area and Price of California Real Estate",
          fontsize = 14,
          weight = 'bold')
plt.xlabel("Area (sq. ft)", weight = 'bold')
plt.ylabel("Price (00's of $)", weight = 'bold')
plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.scatterplot(df_read_estate['Price'],
                df_read_estate['Area (ft.)'],
                hue = df_read_estate['Building Type'],
                palette = ['black', 'darkblue', 'purple', 'pink', 'white'],
                s = 100)
plt.title("Relationship between Area and Price of California Real Estate",
          fontsize = 14,
          weight = 'bold')
plt.xlabel("Area (sq. ft)", weight = 'bold')
plt.ylabel("Price (00's of $)", weight = 'bold')
plt.show()
```

```python
df_ad_expenditure = pd.read_csv(r"C:\Users\mk744\Downloads\365 data science data sets\Budget,Sales.txt")
```

```python
df_ad_expenditure
```

```python
# Linear Regression plot Visualisation
plt.figure(figsize=(10,8))
sns.set(rc = {'figure.figsize':(9,6)})
sns.regplot(x = "Budget",
            y = "Sales",
            data = df_ad_expenditure,
            scatter_kws= {'color': 'red'},
            line_kws= {'color' : 'red'})
plt.xlabel("Ad Expediture in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of Add Expenditure on Sales", fontsize = 14, weight = 'bold')
plt.show()
```

```python
sns.lmplot(x = 'Budget',
           y = 'Sales',
           data = df_ad_expenditure,
           height= 10,
           scatter_kws= {'color': 'red'},
            line_kws= {'color' : 'red'})
plt.xlabel("Ad Expediture in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of Add Expenditure on Sales", fontsize = 14, weight = 'bold')
plt.show()
```
